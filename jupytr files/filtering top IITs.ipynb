```python
df1 = pd.read_csv('2016.csv')
df2 = pd.read_csv('2017.csv')
df3 = pd.read_csv('2018.csv')
df4 = pd.read_csv('2019.csv')
df5 =pd.read_csv('2020.csv')
df6 = pd.read_csv('2021.csv')
df7 = pd.read_csv('2022.csv')
df8 = pd.read_csv('2023.csv')
```
